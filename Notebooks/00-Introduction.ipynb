Inside Jupyter, a *Python3 notebooks* like this one is a standalone document consisting of multiple *cells*. You can 'run' a cell selecting it and either click on the run icon, or press `Shift+Return'.

There are of two main type of cells
* *markdown* cells (like this one) are meant to contain text only. You can use markdown to format text in *italic* or **bold** for instance. When you 'run' a markdown cell, jupyter 'renders' the text. In order to edit a markdown cell, you need to double-click in it
* *code* cells contain actual code, which is executed when you 'run' the cell is executed. Code cells can consist of multiple lines. When you run a code cell, every line is executed, but only the outcome of the last (non-empty) line is printed



```python
1+1
```

You can insert code or markdown cells above or below a given cell using the appropriate button.

## Running python code in Jupyter:
Python code, like most programming language consists in a series of specific commands. Python is *interpreted*, meaning that python code can be executed as is (in contract with *compiled languages* which needs to be converted into a format that a computer can understand).

Python commands are executed sequentially, meaning that you can always count on the fact that line $i$ will be executed before line $i+1$. You can always use `help` or `?` to print documentation on specific python commands.

```python
2+3
print("hello world")
1+1
```

Here, python computed 2+3, printed the text hello world, then computed 1+1. Only the output of the last command (`1+1`) is displayed automatically.
